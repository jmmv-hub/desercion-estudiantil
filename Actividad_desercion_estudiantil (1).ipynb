```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 500

data = pd.DataFrame({
    "age": np.random.randint(16, 30, n),
    "gender": np.random.choice(["Male", "Female"], n),
    "origin": np.random.choice(["Urban", "Rural"], n),
    "highschool_gpa": np.random.uniform(2.0, 5.0, n),
    "admission_score": np.random.uniform(0, 100, n),
    "first_semester_grade": np.random.uniform(0, 5.0, n),
    "socioeconomic_level": np.random.randint(1, 6, n),
    "scholarship": np.random.choice([0,1], n),
    "loan": np.random.choice([0,1], n)
})

# Variable objetivo
data["dropout"] = np.where(
    (data["first_semester_grade"] < 2.5) |
    (data["admission_score"] < 40), 1, 0
)

# Introducir valores nulos
for col in data.columns:
    data.loc[data.sample(frac=0.05).index, col] = np.nan

# Introducir outliers
data.loc[data.sample(frac=0.02).index, "first_semester_grade"] = 10

# Guardar dataset
data.to_csv("student_dropout_dataset.csv", index=False)

print(data.head())

from google.colab import files
files.download("student_dropout_dataset.csv")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

df = data.dropna()

le = LabelEncoder()
df["gender"] = le.fit_transform(df["gender"])
df["origin"] = le.fit_transform(df["origin"])

X = df.drop("dropout", axis=1)
y = df["dropout"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
print("Accuracy:", accuracy)
```
